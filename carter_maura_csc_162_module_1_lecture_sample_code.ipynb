```python
#python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # cv2.imshow() does not work in Colab; this function displays images in Colab.
```

```python
#python
# Read the image file from disk.
#img = cv2.imread('sample_image.jpg')  # Replace 'sample_image.jpg' with your image file path
img = cv2.imread('carter_maura_162_image_reef_fish_school.png')  # Replace 'sample_image.jpg' with your image file path

# OpenCV reads images in BGR format by default.
# Convert the image from BGR to RGB format for correct color display in Colab.
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using cv2_imshow from Google Colab.
cv2_imshow(img_rgb)
```

**4.1 Resizing an Image**
Resizing alters the dimensions of an image either to standardize inputs for a model or to adjust an images size for various applications.

```python
#python
# Resize the image to a fixed width (400 pixels) and height (300 pixels)
resized_img = cv2.resize(img, (400, 300))
cv2_imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
```

**4.2 Cropping an Image**
Cropping lets you select a specific portion of an image and discard the rest. This is useful when you want to focus on a particular region.

```python
#python
# Crop the image. Here, we slice the array: rows 50 to 250 and columns 100 to 400.
cropped_img = img[50:250, 100:400]
cv2_imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
```

**4.3 Rotating an Image**
Rotation changes the orientation of an image, which can correct tilt or create aesthetically rotated views.

```python
#python
# Retrieve the height and width of the image.
(h, w) = img.shape[:2]
# Define the center of the image.
center = (w // 2, h // 2)
# Set the angle of rotation and the scaling factor.
angle = 45  # rotate by 45 degrees
scale = 1.0

# Compute the rotation matrix based on the center point, angle, and scale.
M = cv2.getRotationMatrix2D(center, angle, scale)
# Apply the rotation matrix to perform the affine transformation.
rotated_img = cv2.warpAffine(img, M, (w, h))
cv2_imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
```
